```python
import pandas as pd
```

```python
from scipy.io import arff
```

```python
data = arff.loadarff('4year.arff')
df = pd.DataFrame(data[0])
```

```python
df['bankruptcy'] = (df['class']==b'1')
```

```python
df.head()
```

```python
dataframe_1=pd.DataFrame(df,columns=['Attr1','Attr2','Attr7','Attr10','bankruptcy'])
```

```python
dataframe_1.head()
```

```python
dataframe_1 = dataframe_1.rename(columns={'Attr1':'X1', 'Attr2':'X2','Attr7':'X7','Attr10':'X10'})
```

## Q4 How many companies satisfy the condition, X1 < mean(X1) - stdev(X1) AND X10 < mean(X10) - std(X10)?

```python
condition1 = dataframe_1['X1'] < dataframe_1['X1'].mean() - dataframe_1['X1'].std() 
condition2 = dataframe_1['X10'] < dataframe_1['X10'].mean()- dataframe_1['X10'].std()
```

```python
subset = dataframe_1.loc[condition1&condition2,:]
```

```python
num=subset.bankruptcy.count()
num
```

## Q5 What is the ratio of the bankrupted companies among the sub-groups above?

```python
bankruptcy_ratio=(subset.loc[subset['bankruptcy']]).shape[0]/num
bankruptcy_ratio
```
